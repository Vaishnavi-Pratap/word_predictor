```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
import numpy as np
import pandas as pd
```

```python
# Open the file in read mode
with open('my_sentences_new.txt', 'r') as file:
    # Read the entire contents of the file
    df = file.read()

# Print the contents of the file
print(df)
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([df])
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in df.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=2964)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Embedding

model = Sequential()
model.add(Embedding(2964, 100))
model.add(LSTM(150, return_sequences=True))  # Set return_sequences=True
model.add(LSTM(150))
model.add(Dense(2964, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
import time
text = "what is the"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```

```python
# Save the trained model
model.save('word_generation_model.h5')
```
